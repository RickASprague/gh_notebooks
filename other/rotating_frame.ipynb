# Rotating reference frames

Kinematics of a rotating (non-inertial) frame, built from the single transform $\mathbf p^A = \mathbf R^A_B\,\mathbf p^B + \mathbf t^A$ and differentiated from first principles.

**Contents**
- **Velocity & acceleration** — the *transport theorem*: differentiate the frame transform to get $\mathbf v^A$ and $\mathbf a^A$, and the four-term acceleration (relative / Coriolis / centripetal / Euler), in both inertial-frame (A) and body-frame (B) groupings — including why the same terms are *real* inward accelerations in A but *fictitious* centrifugal / Coriolis forces in B.
- **Nested transforms** ($C\to B\to A$) — a **static mount** (fixed extrinsic $\mathbf R^B_C$ folds into the composition, no new terms), a **dynamic inner spin** (cross-Coriolis and gyroscopic coupling, plus centripetal/Euler on the mounting offset), and **coincident origins** (rotations multiply, angular velocities add — the stack collapses to one rotating frame; the tie to $SE(3)/SE_2(3)$ composition).
- **Worked example** — one horizontal rotating frame read two ways, a **carousel** and a **turning car**, with numbers: a steady turn *is* a carousel, differing only in where you put the origin.

**Conventions:** superscript = destination frame ($\mathbf R^A_B$ maps B→A); $[\cdot]_\times$ = skew/hat operator; $\boldsymbol\omega_{AB}$ = angular velocity of B relative to A, from $\dot{\mathbf R}^A_B = \mathbf R^A_B[\boldsymbol\omega^B_{AB}]_\times$; $\mathbf t^A$ asserted **constant** (a fixed origin of a spinning frame lies on the axis).

# Velocity

$$\begin{aligned}
\mathbf p^A       & = \mathbf R^A_B\, \mathbf p^B + \mathbf t^A \\
\dot{\mathbf p}^A & = \dot{\mathbf R}^A_B\, \mathbf p^B + \mathbf R^A_B\, \dot{\mathbf p}^B \\
\dot{\mathbf p}^A & = \mathbf R^A_B\,(\boldsymbol\omega^B_{AB} \times \mathbf p^B) + \mathbf R^A_B\, \dot{\mathbf p}^B \\
\mathbf v^A       & = \boldsymbol\omega^A_{AB} \times \mathbf p^A + \mathbf R^A_B\, \mathbf v^B \\
\mathbf v^A       & = \mathbf R^A_B\big(\boldsymbol\omega^B_{AB} \times \mathbf p^B + \mathbf v^B\big)
\end{aligned}$$

$\boldsymbol\omega^B_{AB}$ = angular velocity of B **relative to A**, expressed in B (from $\dot{\mathbf R}^A_B=\mathbf R^A_B[\boldsymbol\omega^B_{AB}]_\times$); $\ \boldsymbol\omega^A_{AB}=\mathbf R^A_B\boldsymbol\omega^B_{AB}$ is that same spin in A. $\ \mathbf v^A\equiv\dot{\mathbf p}^A$, $\mathbf v^B\equiv\dot{\mathbf p}^B$.

**Assert $\mathbf t^A$ constant** — it can be *any* fixed offset, **not** necessarily $\mathbf 0$. A fixed origin of a spinning frame lies on the axis, so two things follow: $\dot{\mathbf t}^A=\mathbf 0$ (kills the origin-velocity term), and $\boldsymbol\omega_{AB}\parallel\mathbf t^A\Rightarrow\boldsymbol\omega_{AB}\times\mathbf t^A=\mathbf 0$ (the offset washes out of the transport cross product). Line 5 needs only $\dot{\mathbf t}^A=\mathbf 0$; Line 4's $\boldsymbol\omega^A_{AB}\times\mathbf p^A$ also uses the wash-out.

- **Line 4** groups in the inertial frame A: transport term $\boldsymbol\omega^A_{AB}\times\mathbf p^A$ (the frame dragging the point around) + relative velocity $\mathbf R^A_B\mathbf v^B$.
- **Line 5** factors $\mathbf R^A_B$ out to group everything in the body frame B, $\ \boldsymbol\omega^B_{AB}\times\mathbf p^B + \mathbf v^B\ $ — the velocity twin of the body-frame acceleration form in the next cell.

# Acceleration

Differentiate $\mathbf v^A$ once more ($\mathbf t^A$ **constant**; $\mathbf v^B\equiv\dot{\mathbf p}^B$, $\mathbf a^B\equiv\ddot{\mathbf p}^B$; $\boldsymbol\omega_{AB}$ = angular velocity of B relative to A):

$$\mathbf a^A \equiv \dot{\mathbf v}^A
= \underbrace{\mathbf R^A_B\,\mathbf a^B}_{\text{relative}}
+ \underbrace{2\,\boldsymbol\omega^A_{AB}\times(\mathbf R^A_B\,\mathbf v^B)}_{\text{Coriolis}}
+ \underbrace{\boldsymbol\omega^A_{AB}\times(\boldsymbol\omega^A_{AB}\times\mathbf p^A)}_{\text{centripetal}}
+ \underbrace{\dot{\boldsymbol\omega}^A_{AB}\times\mathbf p^A}_{\text{Euler}}$$

Equivalently, grouped inside the body frame (everything as $\boldsymbol\omega^B_{AB}$, $\mathbf p^B$):

$$\mathbf a^A = \mathbf R^A_B\Big[\,\mathbf a^B + 2\,\boldsymbol\omega^B_{AB}\times\mathbf v^B + \boldsymbol\omega^B_{AB}\times(\boldsymbol\omega^B_{AB}\times\mathbf p^B) + \dot{\boldsymbol\omega}^B_{AB}\times\mathbf p^B\,\Big]$$

Where they come from, differentiating $\dot{\mathbf p}^A=\dot{\mathbf R}^A_B\mathbf p^B+\mathbf R^A_B\dot{\mathbf p}^B$:
- $\ddot{\mathbf R}^A_B\mathbf p^B$ splits into **centripetal** ($\boldsymbol\omega_{AB}\times(\boldsymbol\omega_{AB}\times\mathbf p)$) + **Euler** ($\dot{\boldsymbol\omega}_{AB}\times\mathbf p$),
- the two $\dot{\mathbf R}^A_B\dot{\mathbf p}^B$ cross-terms (one from each product rule) combine into the factor-of-2 **Coriolis** term,
- $\mathbf R^A_B\ddot{\mathbf p}^B$ is the **relative** acceleration.

($\mathbf t^A$ constant ⟹ $\dot{\mathbf t}^A=\ddot{\mathbf t}^A=\mathbf 0$ and $\boldsymbol\omega_{AB}\times\mathbf t^A=\mathbf 0$, so the origin drops out of every term.)

**Sign / naming (the "it's both" point).** As written above — the $\mathbf a^A$ (inertial) form — these are the *real* inward accelerations: **centripetal** points toward the axis, and the $+2\boldsymbol\omega_{AB}\times\mathbf v$ Coriolis term is real. Solve instead for what the rotating observer feels ($\mathbf a^B = \mathbf R^B_A\mathbf a^A - \dots$) and the same two terms cross to the other side with a minus sign, becoming the **fictitious** forces: outward **centrifugal** $-\,\boldsymbol\omega_{AB}\times(\boldsymbol\omega_{AB}\times\mathbf p)$ and $-2\boldsymbol\omega_{AB}\times\mathbf v$. Same physics, opposite sign — which name is "correct" depends on which frame you write the equation for.

# Nested transform ($C\to B\to A$, $\mathbf R^B_C$ static)

$C$ is **rigidly mounted** on $B$: $\mathbf R^B_C,\ \mathbf t^B$ constant $\Rightarrow \dot{\mathbf R}^B_C=0,\ \boldsymbol\omega_{BC}=0$. C has no spin of its own — it's a fixed **extrinsic (mounting)** transform, not a rotating frame.

$$\mathbf p^A = \mathbf R^A_B\,\mathbf p^B + \mathbf t^A,\qquad \mathbf p^B = \mathbf R^B_C\,\mathbf p^C + \mathbf t^B.$$

**Compose:** $\ \mathbf p^A = \mathbf R^A_C\,\mathbf p^C + (\mathbf R^A_B\mathbf t^B + \mathbf t^A),\qquad \mathbf R^A_C\equiv\mathbf R^A_B\mathbf R^B_C.$

## Velocity

The static inner transform only rotates ($\dot{\mathbf R}^B_C=0$, $\dot{\mathbf t}^B=0$), so it passes velocity through unchanged but for the fixed rotation:
$$\mathbf v^B \equiv \dot{\mathbf p}^B = \mathbf R^B_C\,\dot{\mathbf p}^C = \mathbf R^B_C\,\mathbf v^C.$$

Differentiate the outer transform and substitute ($\mathbf t^A$ constant, so $\dot{\mathbf t}^A=0$):
$$\begin{aligned}
\dot{\mathbf p}^A & = \dot{\mathbf R}^A_B\,\mathbf p^B + \mathbf R^A_B\,\dot{\mathbf p}^B \\
                  & = \mathbf R^A_B(\boldsymbol\omega^B_{AB}\times\mathbf p^B) + \mathbf R^A_B\,\mathbf v^B \\
                  & = \mathbf R^A_B(\boldsymbol\omega^B_{AB}\times\mathbf p^B) + \underbrace{\mathbf R^A_B\mathbf R^B_C}_{\mathbf R^A_C}\,\mathbf v^C \\
\mathbf v^A       & = \boldsymbol\omega^A_{AB}\times\mathbf p^A + \mathbf R^A_C\,\mathbf v^C \\
\mathbf v^A       & = \mathbf R^A_B\big(\boldsymbol\omega^B_{AB}\times\mathbf p^B + \mathbf v^B\big)
\end{aligned}$$

Line 4 uses $\mathbf R^A_B(\boldsymbol\omega^B_{AB}\times\mathbf p^B)=\boldsymbol\omega^A_{AB}\times(\mathbf R^A_B\mathbf p^B)=\boldsymbol\omega^A_{AB}\times\mathbf p^A$. Same form as the single-level velocity — only B spins ($\boldsymbol\omega_{AC}=\boldsymbol\omega_{AB}$); the mount just carries the innermost $\mathbf v^C$ out through the composed $\mathbf R^A_C$.

## Acceleration

Static mount again — the inner transform passes acceleration through the fixed rotation too:
$$\mathbf a^B \equiv \ddot{\mathbf p}^B = \mathbf R^B_C\,\ddot{\mathbf p}^C = \mathbf R^B_C\,\mathbf a^C.$$

Differentiate $\mathbf v^A$ once more. The four-term expansion is exactly the single-level one; then push $\mathbf a^B=\mathbf R^B_C\mathbf a^C$ and $\mathbf v^B=\mathbf R^B_C\mathbf v^C$ through $\mathbf R^A_B$ so the inner quantities emerge under the composed $\mathbf R^A_C=\mathbf R^A_B\mathbf R^B_C$:

$$\begin{aligned}
\mathbf a^A & = \mathbf R^A_B\Big[\,\mathbf a^B + 2\,\boldsymbol\omega^B_{AB}\times\mathbf v^B + \boldsymbol\omega^B_{AB}\times(\boldsymbol\omega^B_{AB}\times\mathbf p^B) + \dot{\boldsymbol\omega}^B_{AB}\times\mathbf p^B\,\Big] \\[4pt]
\mathbf a^A & = \underbrace{\mathbf R^A_C\,\mathbf a^C}_{\text{relative}}
+ \underbrace{2\,\boldsymbol\omega^A_{AB}\times(\mathbf R^A_C\,\mathbf v^C)}_{\text{Coriolis}}
+ \underbrace{\boldsymbol\omega^A_{AB}\times(\boldsymbol\omega^A_{AB}\times\mathbf p^A)}_{\text{centripetal}}
+ \underbrace{\dot{\boldsymbol\omega}^A_{AB}\times\mathbf p^A}_{\text{Euler}}
\end{aligned}$$

Same four terms as the single level, with the innermost $\mathbf a^C,\mathbf v^C$ rotated out by $\mathbf R^A_C$ and the spin still just $\boldsymbol\omega_{AB}$.

**No cross-Coriolis.** A coupling term $2\,\boldsymbol\omega_{AB}\times(\boldsymbol\omega_{BC}\times\mathbf p)$ between the two spins would appear *only* if C rotated relative to B ($\boldsymbol\omega_{BC}\neq0$). A static mount has $\boldsymbol\omega_{BC}=0$, so it's absent — the fixed $\mathbf R^B_C$ adds no new fictitious terms, it just folds into $\mathbf R^A_C$.

# Nested transform ($C\to B\to A$, $\mathbf R^B_C$ dynamic)

$\mathbf R^B_C=\mathbf R^B_C(t)$ **can vary with time** — C spins relative to B: $\dot{\mathbf R}^B_C=\mathbf R^B_C[\boldsymbol\omega^C_{BC}]_\times$, $\ \boldsymbol\omega_{BC}\neq0$ (a turntable on the carousel, a gimbal, a joint). Assert $\mathbf t^A,\mathbf t^B$ **constant** — both may be **nonzero** ($\dot{\mathbf t}^A=\dot{\mathbf t}^B=0$).

$$\mathbf p^A = \mathbf R^A_B\,\mathbf p^B + \mathbf t^A,\qquad \mathbf p^B = \mathbf R^B_C\,\mathbf p^C + \mathbf t^B,\qquad \mathbf R^A_C\equiv\mathbf R^A_B\mathbf R^B_C.$$

## Velocity

Differentiate each layer ($\dot{\mathbf t}^A=\dot{\mathbf t}^B=0$; $\dot{\mathbf R}^A_B\mathbf p^B=\mathbf R^A_B(\boldsymbol\omega^B_{AB}\times\mathbf p^B)$, likewise inner):
$$\begin{aligned}
\dot{\mathbf p}^B &= \mathbf R^B_C(\boldsymbol\omega^C_{BC}\times\mathbf p^C + \mathbf v^C) \\
\dot{\mathbf p}^A &= \mathbf R^A_B(\boldsymbol\omega^B_{AB}\times\mathbf p^B) + \mathbf R^A_B\,\dot{\mathbf p}^B \\
\mathbf v^A &= \mathbf R^A_B(\boldsymbol\omega^B_{AB}\times\mathbf p^B) + \mathbf R^A_C(\boldsymbol\omega^C_{BC}\times\mathbf p^C + \mathbf v^C)
\end{aligned}$$

B transports the point about **its own origin**, $\mathbf R^A_B(\boldsymbol\omega^B_{AB}\times\mathbf p^B)$; the $\mathbf R^A_C(\cdot)$ term carries the point's **full velocity as seen in C** — its C-transport $\boldsymbol\omega^C_{BC}\times\mathbf p^C$ **plus** its relative velocity $\mathbf v^C$ — out to A. Equivalently this is the single-level $\mathbf v^A=\mathbf R^A_B(\boldsymbol\omega^B_{AB}\times\mathbf p^B + \mathbf v^B)$ with $\mathbf v^B=\mathbf R^B_C(\boldsymbol\omega^C_{BC}\times\mathbf p^C+\mathbf v^C)$. (The two transports merge into a single $\boldsymbol\omega_{AC}\times\mathbf p^A$ only if the origins coincide.)

## Acceleration

Differentiate once more by **recursion**: the outer single-level formula in $\boldsymbol\omega^B_{AB}$ (valid for constant $\mathbf t^A$), fed by the inner single-level formula for $\mathbf a^B$ in $\boldsymbol\omega^C_{BC}$ (valid for constant $\mathbf t^B$):

$$\mathbf a^A = \mathbf R^A_B\Big[\,
\underbrace{\mathbf a^B}_{\text{relative}}
+ \underbrace{2\,\boldsymbol\omega^B_{AB}\times\mathbf v^B}_{\text{Coriolis}}
+ \underbrace{\boldsymbol\omega^B_{AB}\times(\boldsymbol\omega^B_{AB}\times\mathbf p^B)}_{\text{centripetal}}
+ \underbrace{\dot{\boldsymbol\omega}^B_{AB}\times\mathbf p^B}_{\text{Euler}}\,\Big]$$

$$\mathbf a^B = \mathbf R^B_C\Big[\,
\underbrace{\mathbf a^C}_{\text{relative}}
+ \underbrace{2\,\boldsymbol\omega^C_{BC}\times\mathbf v^C}_{\text{Coriolis}}
+ \underbrace{\boldsymbol\omega^C_{BC}\times(\boldsymbol\omega^C_{BC}\times\mathbf p^C)}_{\text{centripetal}}
+ \underbrace{\dot{\boldsymbol\omega}^C_{BC}\times\mathbf p^C}_{\text{Euler}}\,\Big]$$

with the inner kinematics $\ \mathbf p^B=\mathbf R^B_C\mathbf p^C+\mathbf t^B,\quad \mathbf v^B=\mathbf R^B_C(\boldsymbol\omega^C_{BC}\times\mathbf p^C+\mathbf v^C).$

Substituting $\mathbf a^B,\mathbf v^B,\mathbf p^B$ into the top line sorts the result into three kinds of term:

1. **Each frame's own $\{$relative, Coriolis, centripetal, Euler$\}$ about its own origin** — C's set (in $\boldsymbol\omega^C_{BC}$, about $\mathbf p^C$, all under $\mathbf R^A_C$) from $\mathbf a^B$; B's set (in $\boldsymbol\omega^B_{AB}$, about $\mathbf p^B$).

2. **Cross-coupling of the two spins** — from $2\,\boldsymbol\omega^B_{AB}\times\mathbf v^B$ (B's Coriolis acting on C's transport velocity $\mathbf R^B_C(\boldsymbol\omega^C_{BC}\times\mathbf p^C)$) and, once the two centripetals are referred to a common point, a cross-centrifugal $\boldsymbol\omega_{AB}\times(\boldsymbol\omega_{BC}\times\cdot)+\boldsymbol\omega_{BC}\times(\boldsymbol\omega_{AB}\times\cdot)$ plus the gyroscopic precession $\boldsymbol\omega_{AB}\times\boldsymbol\omega_{BC}$.

3. **Offset (mount) terms from $\mathbf p^B=\mathbf R^B_C\mathbf p^C+\mathbf t^B$** — B's spin also acts on the constant $\mathbf t^B$, giving
$$\boldsymbol\omega^B_{AB}\times(\boldsymbol\omega^B_{AB}\times\mathbf t^B)\quad\text{and}\quad \dot{\boldsymbol\omega}^B_{AB}\times\mathbf t^B,$$
the **centripetal / Euler acceleration of C's mounting point** — C's origin sits at $\mathbf t^B$ off B's axis, so B swings it around. These vanish when $\mathbf t^B=\mathbf 0$; with nonzero constant $\mathbf t^B$ the mount itself accelerates and they stay.

(The clean composite $\mathbf a^A=\mathbf R^A_C\mathbf a^C+2\boldsymbol\omega_{AC}\times(\cdot)+\boldsymbol\omega_{AC}\times(\boldsymbol\omega_{AC}\times\mathbf p^A)+\dot{\boldsymbol\omega}_{AC}\times\mathbf p^A$, $\ \boldsymbol\omega_{AC}=\boldsymbol\omega_{AB}+\boldsymbol\omega_{BC}$, is the **coincident-origin** special case — offsets zero, groups 1–2 merge and group 3 disappears.)

# Nested transform ($C\to B\to A$, coincident origins → $\boldsymbol\omega$ additive)

Take the origins **coincident** ($\mathbf t^A=\mathbf t^B=\mathbf 0$). The offset/mount terms vanish, the two transports merge ($\mathbf R^A_B\mathbf p^B=\mathbf R^A_C\mathbf p^C=\mathbf p^A$), and $C\to A$ becomes a **single rotating frame whose angular velocities add**:

$$\mathbf R^A_C=\mathbf R^A_B\mathbf R^B_C,\qquad \boldsymbol\omega_{AC}=\boldsymbol\omega_{AB}+\boldsymbol\omega_{BC}.$$

The single-level formulas then apply verbatim with the composite $\mathbf R^A_C$ and the summed $\boldsymbol\omega_{AC}$.

## Velocity

$$\mathbf v^A = \boldsymbol\omega^A_{AC}\times\mathbf p^A + \mathbf R^A_C\,\mathbf v^C = \mathbf R^A_C\big(\boldsymbol\omega^C_{AC}\times\mathbf p^C + \mathbf v^C\big)$$

## Acceleration

$$\mathbf a^A = \underbrace{\mathbf R^A_C\,\mathbf a^C}_{\text{relative}} + \underbrace{2\,\boldsymbol\omega^A_{AC}\times(\mathbf R^A_C\mathbf v^C)}_{\text{Coriolis}} + \underbrace{\boldsymbol\omega^A_{AC}\times(\boldsymbol\omega^A_{AC}\times\mathbf p^A)}_{\text{centripetal}} + \underbrace{\dot{\boldsymbol\omega}^A_{AC}\times\mathbf p^A}_{\text{Euler}}$$

Same four-term single-level form, now driven by the **summed** spin $\boldsymbol\omega_{AC}=\boldsymbol\omega_{AB}+\boldsymbol\omega_{BC}$. Expanding that sum re-exposes the two-spin couplings (cross-Coriolis, cross-centrifugal, and the gyroscopic $\boldsymbol\omega_{AB}\times\boldsymbol\omega_{BC}$ inside $\dot{\boldsymbol\omega}_{AC}$) — but they combine cleanly here because both spins share the one origin.

**It keeps recursing:** another coincident layer just adds its spin — $\boldsymbol\omega_{AD}=\boldsymbol\omega_{AB}+\boldsymbol\omega_{BC}+\boldsymbol\omega_{CD}$, and so on. Coincident origins ⟹ rotations multiply, angular velocities add, and the whole stack behaves as one rotating frame.

# Worked example — carousel & turning car

One horizontal rotating frame, read two ways. The spin/turn axis is vertical, $\boldsymbol\omega=\omega\,\hat z$ (the carousel's spin rate, or the car's **yaw rate** $\dot\psi$). From the body-frame form, an object at body position $\mathbf p^B$ moving at $\mathbf v^B$ with relative accel $\mathbf a^B$ has

$$\mathbf a^A=\mathbf R^A_B\big[\underbrace{\mathbf a^B}_{\text{relative}}+\underbrace{2\boldsymbol\omega\times\mathbf v^B}_{\text{Coriolis}}+\underbrace{\boldsymbol\omega\times(\boldsymbol\omega\times\mathbf p^B)}_{\text{centripetal}}+\underbrace{\dot{\boldsymbol\omega}\times\mathbf p^B}_{\text{Euler}}\big].$$

Each bracket term is a **body-frame** acceleration — what you experience in the rotating frame; $\mathbf R^A_B$ merely rotates their sum into A. So we label them $\mathbf a^B_{\text{centripetal}}$, $\mathbf a^B_{\text{Coriolis}}$, etc. (Take $g\approx10\ \mathrm{m/s^2}$ just to gauge magnitudes in "$g$"s.)

## Carousel (origin on the axis)

Body origin at the **center** (on the spin axis). Body axes: $\hat x$ = **radial** (outward), $\hat y$ = **tangential**, $\hat z$ = **up** — coinciding with the spin axis $\boldsymbol\omega$. Constant spin ($\dot{\boldsymbol\omega}=0$). Numbers: $\omega=1\ \mathrm{rad/s}$ ($\approx$ 9.5 rpm), radius $r=2$ m.

**Sitting still** — $\mathbf p^B=(r,0,0)$, $\mathbf v^B=\mathbf a^B=\mathbf 0$:
$$\mathbf a^B_{\text{centripetal}}=\boldsymbol\omega\times(\boldsymbol\omega\times\mathbf p^B)=-\omega^2 r\,\hat x\quad(\text{inward, along }-\hat x,\ \ \omega^2 r=2\ \mathrm{m/s^2}\approx0.2g).$$
Inertially you circle at $\boldsymbol\omega\times\mathbf p=\omega r=2$ m/s (tangential, $\hat y$).

**Walking outward** — $\mathbf v^B=(u,0,0)$, $u=1$ m/s:
$$\mathbf a^B_{\text{Coriolis}}=2\boldsymbol\omega\times\mathbf v^B=2\omega u\,\hat y\quad(\text{tangential, along }\hat y,\ \ 2\omega u=2\ \mathrm{m/s^2}).$$
A $2\ \mathrm{m/s^2}$ tangential deflection while you walk; stop and it vanishes, the centripetal stays.

## Turning car (yaw rate $\omega=v/R$)

A car rounds a curve of radius $R$ at speed $v$; its body frame is $\hat x$ = **forward**, $\hat y$ = **left**, $\hat z$ = **up** (the yaw axis, coinciding with $\boldsymbol\omega$). Note the axes are rotated vs the carousel: here **forward $\hat x$ is tangential**, and **left $\hat y$ is radial** (pointing toward the center on a left turn). It **yaws** at $\omega=\dot\psi=v/R$. Numbers: $v=20$ m/s (72 km/h), $R=100$ m $\Rightarrow\ \omega=0.2\ \mathrm{rad/s}\approx11.5°/\mathrm{s}$.

The body **origin rides with the car** (off the curve's center, radius $R$), so the car's circular motion lives in the moving origin $\ddot{\mathbf t}^A=v^2/R$ toward the center. For a point at the car origin ($\mathbf p^B=\mathbf v^B=\mathbf a^B=\mathbf 0$):
$$\mathbf a^B_{\text{centripetal}}=\frac{v^2}{R}\,\hat y\quad(\text{lateral, toward the turn center},\ \ v^2/R=4\ \mathrm{m/s^2}\approx0.4g).$$

## Same physics, two rides

A steady turn **is** a carousel. The car's lateral $v^2/R$ equals the carousel's $\omega^2 R$ with $R=v/\omega$ — both are just

$$\boxed{\ \mathbf a^B_{\text{centripetal}}=\omega^2\times(\text{distance from the turn center}),\ \text{pointing toward center}.\ }$$

The only difference is bookkeeping — **where you put the origin**:
- **Carousel:** origin **on the axis**, so the centripetal appears as $\boldsymbol\omega\times(\boldsymbol\omega\times\mathbf p^B)$ with $\mathbf p^B=r\,\hat x$.
- **Car:** origin **on the vehicle** (off the center), so the same centripetal appears via the moving origin $\ddot{\mathbf t}^A$. An offset passenger (back seat, $\mathbf p^B\neq\mathbf 0$) picks up an *extra* $\boldsymbol\omega\times(\boldsymbol\omega\times\mathbf p^B)$, and if the driver changes the turn rate, an **Euler** $\dot{\boldsymbol\omega}\times\mathbf p^B$ shove.
