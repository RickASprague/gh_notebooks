# Android IMU

A concrete instance of the body frame $B$: the **Android sensor coordinate system**. With the phone held in its natural (portrait) orientation, looking at the screen:

![Android device sensor axes](../resources/images/android_sensor_axes.png)

| Axis | Points toward | `SensorEvent` |
|---|---|---|
| **+X** | the **right** edge of the screen | `values[0]` |
| **+Y** | the **top** edge (up the screen) | `values[1]` |
| **+Z** | **out of the glass, toward you** (the front face) | `values[2]` |

Right-handed ($X\times Y = Z$), defined relative to the screen's **default** orientation (it does **not** rotate with screen auto-rotation). The accelerometer reports **specific force with gravity included** — i.e. it reads $+9.8$ on whichever axis points up, sensing $-\mathbf g$ (the reaction to gravity), not gravity itself.

**At rest, flat, screen up:** $+Z$ points at the sky, so

$$\texttt{values} \approx (0,\,0,\,+9.81)\ \text{m/s}^2.$$

The $+9.81$ is gravity pushing up against the sensor's proof mass — not motion. Equivalently, *back of the phone pointing down ⟹ screen up ⟹ $+Z$ up ⟹ $z=+9.81$.*

Reference: [Android — Sensors overview](https://developer.android.com/develop/sensors-and-location/sensors/sensors_overview) (`android.hardware.SensorEvent`).
